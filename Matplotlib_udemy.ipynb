````python
years = [1950,1955,1960,1965,1970,1975,1980,1985,
         1990,1995,2000,
         2005,2010,2015]
pops = [2.5,2.7,3,3.3,3.6,4.0,
        4.4,4.8,5.3,5.7,6.1,6.5,6.9,7.3]

deaths = [1.2,1.7,1.8,2.2,2.5,
         2.7,2.9,3,3.1,3.3,3.5,3.8,4.0,4.3 ]
plt.plot(years,pops, color=(255/255,100/255,100/255))
plt.plot(years,deaths)
plt.ylabel("Population in billions")
plt.xlabel("Population growth by year")
plt.title("Population growth")
plt.show()
````

````python
#rather than having 2 plt.plot statements 
#we use a variable to store

lines = plt.plot(years,pops,years,deaths)
plt.setp(lines,color=(1,.4,.4),marker="o")
plt.grid(True)
````

````python
import pandas as pd
data = [{
    'name': "Nick",
    'jan_ir': 124,
    'feb_ir': 100,
    'march_ir': 165
},{
    'name': 'Pandas',
   'jan_ir': 112,
   'feb_ir': 143,
   'march_ir': 3
},
{
    'name': 'Pandas',
   'jan_ir': 112,
   'feb_ir': 143,
   'march_ir': 3
}
]
raw_data = {'names':['Nick','Pandas','S',
                     'Ari','Vales'],
            'jan_ir':[143,122,101,106,365],
            'feb_ir':[122,132,144,98,62],
            'march_ir':[65,88,12,32,65]}
df = pd.DataFrame(raw_data, columns=['names','jan_ir','feb_ir','march_ir'])
print(df,"\n")


df['total_ir']= df['jan_ir']+df['feb_ir']+df['march_ir']
print(df)
````

````python
color = [(1,.4,.4),(1,.6,1),
         (.5,.3,1),
         (.3,1,.5),(.4,.7,.2)]
plt.pie(df['total_ir'],
        labels=df['names'],
        colors=color,
        autopct='%1.1f%%')
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np
col_count=3
bar_width=.1
korea_scores = (554,536,538)
india_scores = (518,523,525)
china_scores = (613,570,580)
france_scores = (495,506,499)

index = np.arange(col_count)
k1 = plt.bar(index,korea_scores,.2,
             label="Korea")
c1 = plt.bar(index + bar_width,china_scores, .1,
             label="China")
f1 = plt.bar(index+0.2,france_scores,.1,
             label = "France")
i1 = plt.bar(index+0.3,india_scores,.1,
             label = "India")

def createLabel(data):
  for item in data:
    height = item.get_height()
    plt.text(item.get_x() + item.get_width()/2.,height*1.05,'%d'%int(height),ha='center',va='bottom')

createLabel(k1)
createLabel(c1)
createLabel(f1)
createLabel(i1)

plt.ylabel("Mean score in PISA 2012")
plt.xlabel("Subjects")
plt.title("Test Scores by the Country")
plt.xticks(index+.3/2,("Mathematics","Reading","Science"))
# plt.legend(edgecolor=None,shadow=False,facecolor=(0,.4,.4))
plt.legend(frameon=False, bbox_to_anchor=(1,1))
# plt.grid(True)
````
